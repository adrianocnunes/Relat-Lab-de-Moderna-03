# Inicializando as bibliotecas

```python
#bibliotecas
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import scipy as sp

%matplotlib inline
%config InlineBackend.figure_format='retina'
```

# Importando a planilha de dados

```python
cm = pd.read_csv('carga_massa4.csv',index_col=0, decimal=',')
```

```python
cm.head()
```

# Mostrando as colunas da planilha

```python
cm.columns
```

# UTILIZANDO OS DADOS PARA r=2

# Definindo as variáveis

```python
x = cm['i^2 (2)'].to_numpy().reshape(-1, 1)
y = cm['VA (V)'].to_numpy().reshape(-1, 1)
```

# Fitando a função e mostrando seus coeficientes

```python
lm = LinearRegression()
lm.fit(x, y)
a = lm.intercept_
b = lm.coef_
print('coeficientes')
print('a=' + str(a))
print('b=' + str(b))
```

# Definindo a função

```python
def f(x):
    return b*x + a
```

# Desenhando o gráfico de VA(V) vs. i^2 (2)

```python
plt.style.use('seaborn-whitegrid')
plt.scatter(x, y, color='red')
plt.xlabel('i^2 (2)')
plt.ylabel('VA (V)')
plt.plot(x, f(x), color='black')
```
